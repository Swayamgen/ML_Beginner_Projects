## Predicting Student Scores Based on Study Hours Using Linear Regression

##### This project aims to predict a student's score based on the number of hours they study per day using a simple linear regression model. Linear regression is a basic yet powerful statistical method used to model the relationship between a dependent variable and one or more independent variables. In this case, the independent variable is the number of hours studied, and the dependent variable is the student's score.

###  Importing Necessary Libraries
##### In this cell, we are importing the necessary libraries:
##### NumPy (np) for numerical operations.
##### Pandas (pd) for data manipulation.
##### LinearRegression from sklearn.linear_model to create and train our linear regression model.
##### Matplotlib (plt) to visualize the data and the regression line.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

### Preparing the Data
##### Explanation:Here, we are creating a dictionary with Hours and Scores, which represents our data. We then convert this dictionary into a pandas DataFrame for easy manipulation and display the first few rows using df.head() to verify that our data is correctly loaded.

```python
data = {
    'Hours': [2.5, 5.1, 3.2, 8.5, 3.5, 1.5, 9.2, 5.5, 8.3, 2.7, 
              7.7, 5.9, 4.5, 3.3, 1.1, 8.9, 2.5, 1.9, 6.1, 7.4, 
              2.7, 4.8, 3.8, 6.9, 7.8],
    'Scores': [21, 47, 27, 75, 30, 20, 88, 60, 81, 25, 
               85, 62, 41, 42, 17, 95, 30, 24, 67, 69, 
               30, 54, 35, 76, 86]
}

df = pd.DataFrame(data)
df.head()  # Displaying the first few rows of the data
```

###  Defining Variables
##### Explanation: In this cell, we define the independent variable X (which represents the hours studied) and the dependent variable y (which represents the scores obtained). This separation is necessary because the model needs to know which variable to use as input and which to predict.

```python
X = df[['Hours']]  # Independent variable: Hours studied
y = df['Scores']   # Dependent variable: Scores obtained
```

### Creating and Training the Model
##### Explanation:Here, we create an instance of the LinearRegression model and train it using the fit() method. The model learns the relationship between the hours studied (X) and the scores obtained (y).

```python
model = LinearRegression()  # Creating the Linear Regression model
model.fit(X, y)             # Training the model on the data
```

### Making a Prediction
##### Explanation:In this cell, we use the trained model to predict the score for a student who studies 9.25 hours per day. The result is stored in predicted_score, and we display it to see the predicted value.

```python
predicted_score = model.predict([[9.25]])  # Predicting the score for 9.25 hours/day
predicted_score[0]
```

### Visualizing the Data and Regression Line
##### Explanation: Here, we create a scatter plot of the actual data points (X vs y) and plot the regression line that the model has learned. This visualization helps in understanding how well the model fits the data.

```python
plt.scatter(X, y, color='blue')              # Plotting the actual data points
plt.plot(X, model.predict(X), color='red')   # Plotting the regression line
plt.title('Hours vs Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.show()
```

### Displaying the Final Predicted Score
##### Explanation: Finally, we print out the predicted score with a formatted message, giving us a clear output of what the model predicts for a student studying 9.25 hours per day.

```python
print(f"The predicted score for a student studying 9.25 hours/day is {predicted_score[0]:.2f}")
```
